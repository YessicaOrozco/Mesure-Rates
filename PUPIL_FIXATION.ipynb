## SOLICITUD 1: DIÁMETRO DE LA PUPILA
Cálculo del diámetro de la pupila dependiendo del área de interés dónde el participante mire.

Se hizo la sumatoria  para los tres experimentos (control, ambiguedad y con la herramienta)  sobre cada área de interés (área de apuestas, el área de información y la pantalla completa dónde se ejecutó el juego).

```python
import csv
with open('MX_metrics_AOI.csv') as file_obj:
    
    reader_obj = csv.reader(file_obj)
    reader_objIJ = list(reader_obj) 
```

```python
print(reader_objIJ[40][15])
```

```python
# Del 0 al 17 are ambiguity // forMX
# Del 18 al 38 are control //for MX
# Del 39 al 65 are tool //for MX

ai = 1
af = 18
ci = 19
cf = 39
hi = 40
hf = 66
```

```python
dpupil_ambg = []
dpupil_control  = []
dpupil_herr  = []

for i in range(ai,af+1):
    if float(reader_objIJ[i][15]) != 0:
        dpupil_ambg.append(float(reader_objIJ[i][15]))
    
#print(dpupil_ambg)
average_ambg = np.average(dpupil_ambg)
sd_ambg = np.nanstd(dpupil_ambg)
var_ambg = np.var(dpupil_ambg)
print('Promedio del diámetro de la pupila durante la ambigüedad: ',average_ambg,)
print('Desviación del diámetro de la pupila durante la ambigüedad: ',sd_ambg,)
print('Varianza del diámetro de la pupila durante la ambigüedad: ',var_ambg,)
print('\n')


for i in range(ci,cf+1):
    #print(float(reader_objIJ[i][15]))
    if float(reader_objIJ[i][15]) != 0:
        dpupil_control.append(float(reader_objIJ[i][15]))

#print(dpupil_control)
average_control = np.average(dpupil_control)
sd_control = np.nanstd(dpupil_control)
var_control = np.var(dpupil_control)
print('Promedio del diámetro de la pupila durante el control: ',average_control,)
print('Desviación del diámetro de la pupila durante el control: ',sd_control,)
print('Varianza del diámetro de la pupila durante el control: ',var_control,)
print('\n')


for i in range(hi,hf+1):
    #print(float(reader_objIJ[i][15]))
    if float(reader_objIJ[i][15]) != 0:
        dpupil_herr.append(float(reader_objIJ[i][15]))

#print(dpupil_herr)
average_herr= np.average(dpupil_herr)
sd_herr = np.nanstd(dpupil_herr)
var_herr = np.var(dpupil_herr)
print('Promedio del diámetro de la pupila durante el uso de la herramienta: ',average_herr,)
print('Desviación del diámetro de la pupila durante el uso de la herramienta: ',sd_herr,)
print('Varianza del diámetro de la pupila durante el uso de la herramienta: ',var_herr,)
print('\n')
```

```python
print(average_ambg)
print(sd_ambg)
print(var_ambg)
print('\n')
print(average_control)
print(sd_control)
print(var_control)
print('\n')
print(average_herr)
print(sd_herr)
print(var_herr)
```
