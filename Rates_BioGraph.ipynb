# MD CONTROL 

```python
# read csv
df_datos1 = pd.read_csv('MD_control.csv')
df_datos1
```

```python
x = df_datos1.iloc[0,2]
x
```

```python
#### ----  MD CONTROL  ---- ######
# Heart rate
b1 = []
for i in range (0,535679):
    x = df_datos1.iloc[i,1]
    b1.append(x)
    
# Skin Conductance
c1 = []
for i in range (0,535679):
    y = df_datos1.iloc[i,2]
    c1.append(y)
    
#Abdominal Respiration
r1 = []
for i in range (0,535679):
    z = df_datos1.iloc[i,3]
    r1.append(z)
```

```python
print('MD CONTROL')
print("\n")

MDcontrol_averageb1 = np.average(b1) # = np.nanmean(EI)
print('El promedio del ritmo cardiaco es: ',MDcontrol_averageb1,)
MDcontrol_sdb1 = np.nanstd(b1)
print('La desviación estandar del ritmo cardiaco es: ',MDcontrol_sdb1,)
MDcontrol_varb1 = np.var(b1)
print('La varianza del ritmo cardiaco es: ',MDcontrol_varb1,)
print("\n")

MDcontrol_averagec1 = np.average(c1) # = np.nanmean(EI)
print('El promedio de la conductancia de la piel es: ',MDcontrol_averagec1,)
MDcontrol_sdc1 = np.nanstd(c1)
print('La desviación estandar de la conductancia de la piel es: ',MDcontrol_sdc1,)
MDcontrol_varc1 = np.var(c1)
print('La varianza de la conductancia de la piel es: ',MDcontrol_varc1,)
print("\n")

MDcontrol_averager1 = np.average(r1) # = np.nanmean(EI)
print('El promedio de la respiración abdominal es: ',MDcontrol_averager1,)
MDcontrol_sdr1 = np.nanstd(r1)
print('La desviación estandar de la respiración abdominal  es: ',MDcontrol_sdr1,)
MDcontrol_varr1 = np.var(r1)
print('La varianza de la respiración abdominal  es: ',MDcontrol_varr1,)
```
